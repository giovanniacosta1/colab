# 游눹 Herramientas

```python
#@title 游볷 Calculadora de Hidrataci칩n
#@markdown Selecciona tu nivel de actividad y tu peso en **LIBRAS**:

actividad = "Intensa" #@param ["Sedentaria", "Moderada", "Intensa"]
peso_en_libras = 196 #@param {type:"slider", min:100, max:300, step:1}

# --- L칩gica Interna (Invisible para el alumno) ---

# 1. Convertimos libras a kilogramos (1 kg = 2.2046 lbs)
peso_kg = peso_en_libras / 2.2046

# 2. Definimos el factor de agua seg칰n actividad
if actividad == "Sedentaria":
  factor = 0.030
elif actividad == "Moderada":
  factor = 0.035
else:
  factor = 0.045

# 3. Calculamos litros
litros = peso_kg * factor

# --- Resultado ---
print("-" * 50)
print(f"游늵 An치lisis para: {peso_en_libras} libras (aprox. {peso_kg:.1f} kg)")
print(f"游끢 Actividad: {actividad}")
print("-" * 50)
print(f"游눦 Meta diaria sugerida: {litros:.2f} LITROS de agua.")
print("-" * 50)
```

# 游깵 Distribuci칩n del Agua


```python
#@title 游늵 Gr치fico de distribuci칩n
import matplotlib.pyplot as plt

# Datos
etiquetas = ['Oc칠anos', 'Glaciares', 'Subterr치nea', 'R칤os/Lagos']
cantidades = [97, 2, 0.7, 0.3]
# Colores visualmente distintos
colores = ['#003f5c', '#7a5195', '#ef5675', '#ffa600']
# "Explotar" (separar) ligeramente las 칰ltimas dos rebanadas peque침as
explosion = (0, 0, 0.1, 0.2)

# Crear la figura
plt.figure(figsize=(8, 6)) # Hacemos la figura un poco m치s ancha para la leyenda

# Generar el gr치fico
# pctdistance=1.1 mueve el porcentaje fuera del c칤rculo (1.0 es el borde)
# labels=None quita los nombres del gr치fico para ponerlos en leyenda
plt.pie(cantidades,
        autopct='%1.1f%%',
        colors=colores,
        explode=explosion,
        shadow=True,
        startangle=45,
        pctdistance=1.15,
        labels=None) # Quitamos las etiquetas directas

plt.title('Distribuci칩n del Agua en la Tierra', y=1.05, fontsize=14)
plt.axis('equal') # Asegura que el pastel sea un c칤rculo perfecto

# A침adir una leyenda a la derecha
plt.legend(labels=etiquetas,
           title="Reservas de Agua",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1)) # Posiciona la leyenda fuera del gr치fico

plt.show()
```
